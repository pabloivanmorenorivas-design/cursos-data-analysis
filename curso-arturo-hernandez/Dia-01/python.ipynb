### **2. ESTRUCTURAS DE DATOS**

#### **2.1 LISTAS**

```python
################### 1.CREACIÓN DE LISTAS #####################

# 1. Lista vacía
lista_vacia = list()
lista_vacia_2 = []

# 2. Lista con elementos
motocicletas = ["honda", "yamaha", "suzuki", "yamaha"]
print(motocicletas)

################### 2.INSERCIONES Y ACTUALIZACIONES #####################
motocicletas.append("ducati")
motocicletas.insert(0, "daelim")
motocicletas[1] = "bmw"
print(motocicletas)

################### 3.BORRADOS #####################
elemento_cero = motocicletas.pop(0)
ultimo_elemento = motocicletas.pop()
print(motocicletas)

motocicletas.remove("yamaha")
print(motocicletas)

motocicletas.clear()
print(motocicletas)

################# 4.ACCESO A LAS LISTAS ######################

# La forma de acceder a los elementos de una lista es con el operador [inicio:fin]. Indicando el inicio de elementos que quieres de la lista
# y fin indica el elemento - 1 que va recoger de la lista. Los indices empiezan en 0.
numeros = [1, 2, 3, 4, 5]
primer_elemento = numeros[0]
dos_primeros_elementos = numeros[:2]
print(dos_primeros_elementos)
```

#### **2.2 SETS**

```python
################### 1.CREACIÓN DE SETS #####################

# 1. set vacío
nuevo_set = set()

# 2. Inicializar un set con elementos
nuevo_set = {1, 2, 3, 2}
print(nuevo_set)

################### 2.INSERTAR Y ACTUALIZAR #####################
nuevo_set.add(4)
nuevo_set.update({4, 5, 6, 7})
print(nuevo_set)
copia_set = nuevo_set.copy()

################### 3.BORRADOS #####################
nuevo_set.discard(2)
nuevo_set.remove(3) # Lanza error si no exikste el elemento
print(nuevo_set)

################### 4.OPERACIONES DE SETS #####################
set_1 = {1, 2, 3, 4}
set_2 = {4, 5, 6, 7}

set_diferencia = set_1.difference(set_2)
print(set_diferencia)

diferencia_simetrica = set_1.symmetric_difference(set_2)
print(diferencia_simetrica)

interseccion = set_1.intersection(set_2)
print(interseccion)

union = set_1.union(set_2)
print(union)

################### 5.ACCESO A LOS SETS #####################
# Solo se puede acceder con el for loop
for element in union:
    print(element)

#No hay una forma de acceder de forma indexada

```

#### **2.3 TUPLAS**

```python
################### 1.CREACIÓN DE TUPLAS #####################
tupla_vacia = ()
tupla_vacia = tuple()

################### 2.INSERTAR Y ACTUALIZAR #####################
tupla_uno = ("Pablo", "Ivan", "Moreno", "Rivas")
tupla_dos = ("r2d2", "coder")

################### 3.BORRADOS #####################
del tupla_vacia

################### 4.OPERACIONES DE TUPLAS #####################
tupla_tres = tupla_uno + tupla_dos
print(tupla_tres)
index_pablo = tupla_uno.index("Pablo")
print(index_pablo)
count_coder = tupla_dos.count("coder")
print(count_coder)

################### 5.ACCESO A LOS TUPLAS #####################

# Acceso de forma indexada
print(tupla_uno[0])

# Acceso con intervalos de elementos
print(tupla_dos[-2:])

# Acceso con bucle for
for elemento in tupla_uno:
    print(elemento)
```

#### **2.4 DICCIONARIOS**

```python
################### 1.CREACIÓN Y ACCESO DE DICCIONARIOS #####################

# Diccionario vacío inicializado con dict()
diccionario_vacio = dict()
diccionario_vacio = {}

# Diccionario definido con tuplas
diccionario = dict([("clave1", "valor1"), ("clave2", "valor2")])
print(diccionario)

# Diccionario inicializado con valores
tecnologias = {"Web": ["HTML", "CSS", "JavaScript"], "Videojuegos": ["Unity", "C#"]}
print(tecnologias)

################### 2.INSERCION Y ACTUALIZACION EN DICCIONARIOS #####################
tecnologias["Data Engineer"] = ["Python", "Hadoop"]
print(tecnologias)

tecnologias.update({"Desarrollo Movil": ["Android", "iOS"], "Web": ["HTML", "CSS", "React", "JavaScript"]})
print(tecnologias)

tecnologias["Web"].append("TailwindCSS")
print(tecnologias)

################### 3.BORRADOS DE ELEMENTOS EN DICCIONARIOS #####################
valor_borrado = tecnologias.pop("Data Engineer")
print(tecnologias)
print(valor_borrado)

# tecnologias.clear()
# print(tecnologias)

################### 4.ACCESO A ELEMENTOS EN DICCIONARIOS #####################
tecnologias_web = tecnologias["Web"]
tecnologia_data_analyst = tecnologias.get("Data Analyst", "No existe en el diccionario")
print(tecnologias_web)
print(tecnologia_data_analyst)

# Obtener todos los objetos del diccionario
print(tecnologias.items())

# Obtener todas las llaves del diccionario
print(tecnologias.keys())

# Obtener todos los valores del diccionario
print(tecnologias.values())

# iterador
for key, value in tecnologias.items():
    print(key + ":", value)
```
