# Indexación Lógica

```python
# Sección de importación de numpy
import numpy as np

# Generamos la semilla para la reproducibilidad del código
np.random.seed(59)

# Creamos el array del banner publicitario con 60 valores de 100 a 10000
ad_clicks = np.random.randint(100, 10001, 60, np.int32)
```

```python
# Mostramo el array en pantalla
print(ad_clicks)
```

```python
# Muestra todos los elementos del array donde el número de clics sea mayor que 8000
greater_than_8000 = ad_clicks > 8000
gt_8000_array = ad_clicks[greater_than_8000]

print(f"Estos son los elementos del array mayores a 8000:\n{gt_8000_array}")
```

```python
# Muestra el recuento de dichos valores
total_gt_8000 = gt_8000_array.size

print(f"Hay un total de {total_gt_8000} días con más de 8000 clics")
```

```python
# Calcula el porcentaje de valores y muestra los resultados en la pantalla:
# Más de 4000 clics:
more_than_4000 = ad_clicks[ad_clicks > 4000].size / ad_clicks.size * 100
print(f"El {more_than_4000:.2f}% de los días tuvieron más 4000 clics.")

# No más de 1000 clics:
less_equal_to_1000 = ad_clicks[ad_clicks <= 1000].size / ad_clicks.size * 100
print(f"\nEl {less_equal_to_1000:.2f}% de los días tuvieron 1000 clics o menos.")

# Entre 5000 y 7500 clics:
btwn_5000_7500_clicks = ad_clicks[np.logical_and(ad_clicks >= 5000, ad_clicks <= 7500)].size / ad_clicks.size * 100
print(f"\nEl {btwn_5000_7500_clicks:.2f}% de los valores tuvieron entre 5000 y 7500 clics.")
```
