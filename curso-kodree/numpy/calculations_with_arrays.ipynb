# Cálculos con Arrays de NumPy

```python
# Sección de importación de numpy
import numpy as np

# Generamos la semilla para la reproducibilidad del código
np.random.seed(59)

# Creamos el array del banner publicitario con 60 valores de 100 a 10000
ad_clicks = np.random.randint(100, 10001, 60, np.int32)
```

```python
# Mostramos array en pantalla
print(ad_clicks)
```

```python
# Encuentra el promedio y la mediana del número de clics en el banner publicitario
mean_clicks = np.mean(ad_clicks)
print(f"Hay un promedio de {mean_clicks:.2f} clics por día durante la campaña publicitaria")

median_clicks = np.median(ad_clicks)
print(f"La mediana del banner es de {median_clicks} clics")
```

```python
# Calcula el número total de clics para los primeros 30 días y los últimos 30 días
first_30_days = np.sum(ad_clicks[:31])
last_30_days = np.sum(ad_clicks[-30:])

print(f"El primer mes hubo un total de {first_30_days} clics en el banner")
print(f"El último mes hubo un total de {last_30_days} clics en el banner")

# Determina el mes más exitoso
if first_30_days > last_30_days:
    print("El primer mes tuvo un mayor alcance")
elif last_30_days > first_30_days:
    print("El segundo mes tuvo un mayor alcance")
else:
    print("Ambos meses tuvieron el mismo nivel de alcance")
```

```python
# Encuentra el número máximo y mínimo de clics
max_clicks = np.max(ad_clicks)
min_clicks = np.min(ad_clicks)

print(f"El número máximo de clics fue de {max_clicks} y el número mínimo de clics fue de {min_clicks}")
```

```python
# Calcula el rango de clics
clicks_range = max_clicks - min_clicks
print(f"Tenemos un rango de {clicks_range} clics")
```

```python
# Supón que el costo por clic en el anuncio es de $0.03
# Calcula el costo del anuncio para cada día, así como el costo total del anuncio

cost_per_click = 0.03
total_clicks = np.sum(ad_clicks)
total_cost = total_clicks * cost_per_click

print(f"Con un total de {total_clicks} clics el anuncio tuvo un costo total de ${total_cost}")
```
