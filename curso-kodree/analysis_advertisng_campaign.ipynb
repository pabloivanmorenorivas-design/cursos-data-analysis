# Analysis of the advertising campaign

```python
# Importa NumPy
import numpy as np
```

```python
# Antes que naada como usaremos random, definiremos una semilla para que se pueda replicar el resultado
np.random.seed(123456)

# Crear array publicitario aleatorio con 60 elementos del 100 al 10000 que representen los clics por día en 60 días
ad_clicks = np.random.randint(low=100, high=10001, size=60, dtype=np.int32)
```

```python
# Mostrar el array en pantalla
print(ad_clicks)
```

```python
# Contar número de elementos en el array
array_size = ad_clicks.size
print(f"Hay un total de {array_size} elementos.")
```

```python
# Determinar tipo de dato del array de numpy
array_type = ad_clicks.dtype
print(f"Los datos del banner publiciatrio son de tipo {array_type}")
```
