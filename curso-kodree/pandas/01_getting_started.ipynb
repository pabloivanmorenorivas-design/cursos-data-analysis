# Conclusiones del Análisis Exploratorio (EDA)

### 1. Dimensión del Dataset
El conjunto de datos cuenta con un total de **51,290 filas y 16 columnas**, lo que proporciona un volumen sólido de registros para el análisis de ventas.

### 2. Tipos de Datos y Consistencia
* **Variables numéricas forzadas:** Las columnas se reconocieron en su mayoría de forma automática. Sin embargo, campos como `aging` y `quantity` se identificaron como decimales (`float64`) debido a la presencia de valores nulos (`NaN`). En un entorno de producción, lo ideal sería transformarlas a enteros (`int64`) tras realizar la limpieza.
* **Variables temporales:** Las columnas de fecha (`order_date`) y hora (`time`) se cargaron como objetos de texto (`object`). Para análisis temporales avanzados, se recomienda combinarlas y convertirlas al tipo `datetime`.

### 3. Valores Faltantes
Se detectó una cantidad mínima de valores faltantes distribuidos en las siguientes variables:
* `aging` (1)
* `sales` (1)
* `quantity` (2)
* `discount` (1)
* `shipping_cost` (1)
* `order_priority` (2)

Al representar menos del **0.01%** del total de los datos, no afectan significativamente el sesgo del dataset y pueden ser fácilmente imputados o removidos durante la fase de limpieza.

### 4. Valores Inusuales o Anómalos Detectados
* **Anomalía en `aging`:** Al inspeccionar los valores únicos de la columna mediante `.unique()`, se identificó un registro aislado con el valor de `10.5`, mientras que el resto de los datos son números enteros estrictos del 1 al 10. Esto se diagnostica como un error de captura o ruido didáctico.
* **Anomalía en `shipping_cost`:** El análisis descriptivo reveló un costo de envío mínimo de **0.10 dólares**. Considerando que la media es de `7.04` dólares y el 75% supera los `2.50` dólares, esta tarifa resulta sospechosamente baja y podría representar un error de registro o una promoción específica no estandarizada.
* **Ruido en Identificadores:** El método `.describe()` procesó la columna `customer_id` por ser numérica, generando promedios y desviaciones estándar que carecen de valor analítico al tratarse de una variable categórica de identificación.
