```python
# Sección de importación
import numpy as np

# Como trabajaremos con np.random fijaremos una semilla para garantizar la reproducibilidad del código
np.random.seed(59)

# Crear el array ad_clicks que contenga 60 elementos aleatorios entre 100 y 10000
ad_clicks = np.random.randint(100, 10001, 60, np.int64)
```

```python
# Mostramos el array en la pantalla
print(ad_clicks)
```

```python
# Muestra el número de clics:
# Del primer día:
first_day_clicks = ad_clicks[0]
print(f"El primer día hubo un total de {first_day_clicks} clics en el banner.")

# Del último día:
last_day_clicks = ad_clicks[-1]
print(f"El último día hubo un total de {last_day_clicks} clics en el banner.")

# Durante los primeros 10 días:
first_10_days = np.sum(ad_clicks[:10])
print(f"Hubo un total de {first_10_days} clics en el banner durante los primeros 10 días.")

# Durante el último mes:
last_month = np.sum(ad_clicks[-30:])
print(f"El último mes hubo un total de {last_month} clics en el banner.")
```
